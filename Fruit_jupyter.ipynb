## converting the data

```python
df['Fruit'] = df['Fruit'].astype('category')
df['Fruit'] = df['Fruit'].cat.codes

#df['Form'] = df['Form'].astype('category')
#df['Form'] = df['Form'].cat.codes

df['RetailPriceUnit'] = df['RetailPriceUnit'].astype('category')
df['RetailPriceUnit'] = df['RetailPriceUnit'].cat.codes

df['CupEquivalentUnit'] = df['CupEquivalentUnit'].astype('category')
df['CupEquivalentUnit'] = df['CupEquivalentUnit'].cat.codes
```
